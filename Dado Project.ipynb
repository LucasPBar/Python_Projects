Crie um programa em Python, que simule lançamentos de dados. 
Sorteie aleatóriamente um número dentre 1 e 6.

Depois sorteie novamente um número dentre 1 e 6 multiplique-os e armazene, obviamente, os resultados serão entre 1 e 36. 

I)  Crie um gráfico para visualizar a distribuição de números sorteados para apenas um Dado com sorteios de 1 a 6.

II)  Crie um gráfico para comparação do lançamentos dos dois dados de 1 a 6.

III)  Crie por fim, um gráfico demonstrando a distribuição dos lançamentos do produto dos resultados dos sorteio dos dois dados.

IV) Crie por fim, um gráfico demonstrando a distribuição dos lançamentos da soma dos resultados dos sorteio dos dois dados.

Faça 3 simulações:<br>

100 lançamentos<br>
1.000 lançamentos<br>
10.000 lançamentos

```python
"""Importação de blibiotecas"""

import random
import matplotlib.pyplot as plt
```

```python
"""Numero de lançamentos"""
lancamento = 1000
```

```python
# list_compreention = [operacao for x in lista]


intervalo_multiplicacao = [c * i for c in range(1, 7) for i in range(1,7)] # Sortear dois valores em um intervalo de até 6 e vai multiplicar os valores obtidos
intervalo_multiplicacao = list(set(intervalo_multiplicacao)) # Listar os valroes únicos
intervalo_multiplicacao.sort() #Ordenação

#lista = []
#for x in range(1,7):
#    for i range(1,7):
#    lista.append(x * i)


"""Sorteio dos valores"""
dado1 = [random.randint(1,6) for c in range(lancamento)]
dado2 = [random.randint(1,6) for c in range(lancamento)]
multi = [dado1[c] * dado2[c] for c in range(lancamento)]
soma = [dado1[c] + dado2[c] for c in range(lancamento)]

"""Contagem dos dados obtidos"""
intervalo1 = [{c:dado1.count(c)} for c in range(1,7)]
intervalo2 = [{c:dado2.count(c)} for c in range(1,7)]
intervalo3 = [{c:multi.count(c)} for c in intervalo_multiplicacao]
intervalo4 = [{c:soma.count(c)} for c in range(2,13)]

print(intervalo1)
print(intervalo2)
print(intervalo3)
print(intervalo4)


```

- I)  Crie um gráfico para visualizar a distribuição de números sorteados para apenas um Dado com sorteios de 1 a 6.

```python
"""Dado 01"""
intervalo = intervalo1
x = [list(i.keys())[0] for i in intervalo]
y = [list(i.values())[0] for i in intervalo]

#for x in intervalo:
#   print(x.keys())

"""Formatando Gráfico"""

plt.bar(x,y,color='orange')
plt.title(f'Comparação de {lancamento} lançamentos entre dois dados (Dado 01)')
plt.xlabel("Face")
plt.ylabel("Frequência")
plt.ylim(0, max(y) + max(y)*0.1)  # Ajusta o eixo Y para começar em zero até um pouco acima do valor máximo de y
plt.grid(True, axis='y')  # Grade apenas no eixo Y para visualização mais clara
for i in range(len(x)):
    plt.text(x[i], y[i], f"{y[i]}", ha='center', va='bottom')  # Centralizar rótulo em cima da barra
plt.show()
```

```python
"""Dado 02"""
intervalo = intervalo2
x = [list(i.keys())[0] for i in intervalo]
y = [list(i.values())[0] for i in intervalo]

#for x in intervalo:
#   print(x.keys())

"""Formatando Gráfico"""

plt.bar(x,y,color='orange')
plt.title(f'Comparação de {lancamento} lançamentos entre dois dados (Dado 02)')
plt.xlabel("Face")
plt.ylabel("Frequência")
plt.ylim(0, max(y) + max(y)*0.1)  # Ajusta o eixo Y para começar em zero até um pouco acima do valor máximo de y
plt.grid(True, axis='y')  # Grade apenas no eixo Y para visualização mais clara
for i in range(len(x)):
    plt.text(x[i], y[i], f"{y[i]}", ha='center', va='bottom')  # Centralizar rótulo em cima da barra
plt.show()
```

```python
# Obs: Perceba que quando você soma duas ou mais distribuições uniformes, o resultado se assemelha a uma distribuição normal

intervalo = intervalo4
x = [list(i.keys())[0] for i in intervalo]
y = [list(i.values())[0] for i in intervalo]

#for x in intervalo:
#   print(x.keys())

"""Formatando Gráfico"""

plt.bar(x,y,color='orange')
plt.title(f'Comparação de {lancamento} lançamentos entre dois dados')
plt.xlabel("Face")
plt.ylabel("Frequência")
plt.ylim(0, max(y) + max(y)*0.1)  # Ajusta o eixo Y para começar em zero até um pouco acima do valor máximo de y
plt.grid(True, axis='y')  # Grade apenas no eixo Y para visualização mais clara
for i in range(len(x)):
    plt.text(x[i], y[i], f"{y[i]}", ha='center', va='bottom')  # Centralizar rótulo em cima da barra
plt.show()
```

### Insights obtidos

1. **Distribuição Uniforme Esperada**: Em um dado justo, cada face deve ter uma probabilidade de 16≈16.67% 61≈16.67%, resultando em aproximadamente 166.67 ocorrências em 1000 lançamentos.

2. **Lei dos Grandes Números**: Com um grande número de lançamentos, as frequências observadas devem convergir para as probabilidades teóricas, indicando uma distribuição uniforme.

3. **Variância e Desvio Padrão**: A variância mede a dispersão dos dados; uma variância alta pode indicar que os dados não são justos.

4. **Distribuição Assimétrica do Produto**: O produto de dois dados resulta em uma distribuição assimétrica, ou seja, o número que obtiver mais múltiplos é o que vai aparecer com mais frequência.

5. **Independência dos Dados**: A correlação entre os resultados dos dois dados deve ser próxima de zero para garantir que sejam independentes.

6. **Intervalos de Confiança**: Fornecem uma faixa estimada para a verdadeira probabilidade de cada face ou soma, ajudando a identificar possíveis vieses.

7. **Repetibilidade e Consistência**: Realizar múltiplas simulações ajuda a confirmar a aleatoriedade e a consistência dos resultados, garantindo que a simulação não é inviezada.
